```python

plt.figure(figsize=(10, 6))
data['Continent'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Categorical Variable Distribution')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.hist(data['Growth Rate'], bins=20, color='orange', edgecolor='black')
plt.title('Continuous Variable Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python

corr_matrix = data.corr()
print("Correlation matrix:\n", corr_matrix)
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(data['Continent'], data['Growth Rate'], alpha=0.5, color='green')
plt.title('Scatter Plot: X vs. Y')
plt.xlabel('Continent')
plt.ylabel('Growth Rate')
plt.show()
```

```python
# Example: Distribution of a target variable
plt.figure(figsize=(6, 4))
data['Density (per km²)'].value_counts().plot(kind='bar', color='purple')
plt.title('Target Variable Distribution')
plt.xlabel('Target')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```
